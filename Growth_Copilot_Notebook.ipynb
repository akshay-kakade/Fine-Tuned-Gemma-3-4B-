### 🧠 Step 2: Model Initialization
Loading the Gemma 3 4B base model with 4-bit quantization and configuring the LoRA adapters via Unsloth.

```python
from unsloth import FastLanguageModel
import torch

max_seq_length = 2048
dtype = None # Auto-detects based on system (FP16 for T4)
load_in_4bit = True # Enforces memory optimization

# 1. Clear out physical memory leftovers
torch.cuda.empty_cache()

# 2. Base model instantiation
model, tokenizer = FastLanguageModel.from_pretrained(
    model_name = "unsloth/gemma-3-4b-it-bnb-4bit", # Official instruction-tuned model asset
    max_seq_length = max_seq_length,
    dtype = dtype,
    load_in_4bit = load_in_4bit,
)

# 3. Injecting PEFT (LoRA) Targets across the complete attention mapping
model = FastLanguageModel.get_peft_model(
    model,
    r = 16,
    target_modules = ["q_proj", "k_proj", "v_proj", "o_proj", "gate_proj", "up_proj", "down_proj"],
    lora_alpha = 16,
    lora_dropout = 0,
    bias = "none",
    use_gradient_checkpointing = "unsloth",
    random_state = 3407,
    use_rslora = False,
    loftq_config = None,
)
```

### 📝 Step 3: Prompt Formatting
Applying the ChatML/Gemma template to ensure the conversational data matches the model's training structure.

```python
from unsloth.chat_templates import get_chat_template

# Map standard conversational arrays into Gemma-3 structural formatting
tokenizer = get_chat_template(
    tokenizer,
    chat_template = "gemma", # Forces target template guidelines
    mapping = {"role" : "from", "content" : "value", "user" : "human", "assistant" : "gpt"},
)

def formatting_prompts_func(examples):
    convs = examples["conversations"]
    texts = [tokenizer.apply_chat_template(convo, tokenize = False, add_generation_prompt = False) for convo in convs]
    return { "text" : texts }

# Apply tokenization mappings across the processing pool
hf_dataset = hf_dataset.map(formatting_prompts_func, batched = True)
```

### ⚡ Step 4: Supervised Fine-Tuning (SFT)
Running the trainer with optimized hyperparameters for the Tesla T4 GPU.

```python
from trl import SFTTrainer
from transformers import TrainingArguments

trainer = SFTTrainer(
    model = model,
    tokenizer = tokenizer,
    train_dataset = hf_dataset,
    dataset_text_field = "text",
    max_seq_length = max_seq_length,
    dataset_num_proc = 2,
    args = TrainingArguments(
        per_device_train_batch_size = 2,
        gradient_accumulation_steps = 4,
        warmup_steps = 5,
        max_steps = 60,              # Short test block to verify training trajectory instantly
        learning_rate = 2e-4,         # Standard, highly effective Unsloth learning rate threshold
        fp16 = True,                  # Mandatory hardware precision for T4 GPU compatibility
        bf16 = False,                 # Prevent NaN generation loops
        logging_steps = 1,            # Real-time loss tracing per step
        optim = "adamw_8bit",         # Standard production quantization optimizer
        weight_decay = 0.01,
        lr_scheduler_type = "linear",
        seed = 3407,
        output_dir = "outputs",
        save_strategy = "no",
        max_grad_norm = 1.0,          # Standard bounding layer to eliminate gradient spikes
    ),
)

# Kick off your new stable run!
trainer_stats = trainer.train()
```
